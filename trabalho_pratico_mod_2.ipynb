```python
# quantas acoes tem 1,2,3,4 e 5 caracteres em seu nome, respectivamente?
acoes = df1.groupby('Name').count().reset_index()
um = 0
dois = 0
tres = 0
quatro = 0
cinco = 0

for i in acoes['Name']:
  if len(i) == 1:
    um += 1
  elif len(i) == 2:
    dois += 1
  elif len(i) == 3:
    tres += 1
  elif len(i) == 4:
    quatro += 1
  elif len(i) == 5:
    cinco += 1
print(um)
print(dois)
print(tres)
print(quatro)
print(cinco)
```
